### Figure 1

```python
# Objective Function
def f(x):
    return np.sum([np.abs(x - i/100) + np.abs(x + i/100) - 2*i/100 for i in range(99)])

x_0 = -10.0
x_end = 1.0
X = np.linspace(x_0, x_end, 500)       
Y = np.array([f(x) for x in X])            

plt.figure(figsize=(5, 3))
plt.plot(X, Y, label='$f(x)$')
plt.title('The objective function')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
step_size = specular.StepSize(name='not_summable', parameters=1.0)
specular.gradient_method(f=f, x_0=x_0, step_size=step_size, form='implicit')
```
