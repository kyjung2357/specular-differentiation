Example 1

```python
t_0 = 0.0
T = 2.5
h = 0.1
max_iter = 1000

def F(t, u):
    return -3*u 

def exact_sol(t):
    return np.exp(-3*t)

def u_0(t_0):
    return exact_sol(t_0)

result_explicit = Explicit_Euler_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h)
result_implicit = Implicit_Euler_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h)
result_CN = Crank_Nicolson_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h)
result_ST = Specular_Trigonometric_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h, u_1=exact_sol(t_0 + h))

exact_solution = np.array([exact_sol(t) for t in result_explicit[0]])

plt.figure(figsize=(5.5, 2.5))
           
plt.plot(result_explicit[0], exact_solution, color='black', label='exact solution') 

plt.plot(result_explicit[0], result_explicit[1],  marker='x', linestyle='None',
         markerfacecolor='none', 
         markeredgecolor='red',  
         label='Explicit Euler') 

plt.plot(result_implicit[0], result_implicit[1],  marker='x', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='blue',  
         label='Implicit Euler')  

plt.plot(result_CN[0], result_CN[1],  marker='x', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='purple',
         label='Crank-Nicolson')  

plt.plot(result_ST[0], result_ST[1],  marker='o', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='green',  
         label='Specular trigonometric')

plt.xlabel(r"time $t$", fontsize=10)
plt.ylabel(r"solution $u(t)$", fontsize=10)
plt.grid(True)
plt.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), borderaxespad=0., fontsize=10)
plt.savefig('figures/Numerical example 1.png', dpi=1000, bbox_inches='tight')
plt.show()
```

```python
t_0 = 0.0
T = 0.9
h = 0.01
max_iter = 1000

def F(t, u):
    return -(t*u)/(1-t**2)

def exact_sol(t):
    return np.sqrt(1 - t**2)
    
def u_0(t_0):
    return exact_sol(t_0)

result_explicit = Explicit_Euler_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h)
result_implicit = Implicit_Euler_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h)
result_CN = Crank_Nicolson_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h)
result_5 = Specular_Euler_Scheme_Type_5(F, u_0=u_0, t_0=t_0, T=T, h=h, max_iter=max_iter)
result_6 = Specular_Euler_Scheme_Type_6(F, u_0=u_0, t_0=t_0, T=T, h=h, max_iter=max_iter)

exact_solution = np.array([exact_sol(t) for t in result_implicit[0]])

plt.figure(figsize=(5.5, 2.5))
           
plt.plot(result_explicit[0], exact_solution, color='black', label='exact solution') 

plt.plot(result_explicit[0], result_explicit[1],  marker='x', linestyle='None',
         markerfacecolor='none', 
         markeredgecolor='red',  
         label='Explicit Euler') 

plt.plot(result_implicit[0], result_implicit[1],  marker='x', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='blue',  
         label='Implicit Euler')  

plt.plot(result_CN[0], result_CN[1],  marker='x', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='purple',
         label='Crank-Nicolson')  

plt.plot(result_5[0], result_5[1],  marker='v', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='green',
         label='Specular Euler (Type 5)')

plt.plot(result_6[0], result_6[1],  marker='v', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='orange',  
         label='Specular Euler (Type 6)')

plt.xlabel(r"time $t$", fontsize=10)
plt.ylabel(r"solution $u(t)$", fontsize=10)
plt.grid(True)
plt.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), borderaxespad=0., fontsize=10)
plt.savefig('figures/Numerical example 5-1.png', dpi=1000, bbox_inches='tight')
plt.show()

```

```python
error_list_explicit = []
error_list_implicit = []
error_list_CN = []
error_list_5 = []
error_list_6 = []


for k in range(3, 18):
    n = 2**k
    h = 1/n

    print(k)
    
    result_explicit = Explicit_Euler_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h)
    result_implicit = Implicit_Euler_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h)
    result_CN = Crank_Nicolson_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h)
    result_5 = Specular_Euler_Scheme_Type_5(F, u_0=u_0, t_0=t_0, T=T, h=h, max_iter=max_iter)
    result_6 = Specular_Euler_Scheme_Type_6(F, u_0=u_0, t_0=t_0, T=T, h=h, max_iter=max_iter)

    exact_solution = np.array([exact_sol(t) for t in np.arange(t_0, T + h, h)])
    
    length = len(result_explicit[1])
    
    error_list_explicit.append((n, total_error(n, exact_solution[:length], result_explicit[1], norm='max')))
    error_list_implicit.append((n, total_error(n, exact_solution[:length], result_implicit[1], norm='max')))
    error_list_CN.append((n, total_error(n, exact_solution[:length], result_CN[1], norm='max')))
    error_list_5.append((n, total_error(n, exact_solution[:length], result_5[1], norm='max')))
    error_list_6.append((n, total_error(n, exact_solution[:length], result_6[1], norm='max')))
    clear_output(wait=True)

compute_ratios(error_list_explicit)
compute_ratios(error_list_implicit)
compute_ratios(error_list_CN)
compute_ratios(error_list_5)
compute_ratios(error_list_6)

clear_output(wait=True)
```

```python
df_explicit = pd.DataFrame(compute_ratios(error_list_explicit), columns=["n", "Error", "Ratio"])
df_implicit = pd.DataFrame(compute_ratios(error_list_implicit), columns=["n", "Error", "Ratio"])
df_CN = pd.DataFrame(compute_ratios(error_list_CN), columns=["n", "Error", "Ratio"])
df_5 = pd.DataFrame(compute_ratios(error_list_5), columns=["n", "Error", "Ratio"])
df_6 = pd.DataFrame(compute_ratios(error_list_6), columns=["n", "Error", "Ratio"])

save_table_to_txt(df_explicit, "tables/5-2 - Explicit Euler Table.txt", error_precision=2, ratio_precision=3)
save_table_to_txt(df_implicit, "tables/5-2 - Implicit Euler Table.txt", error_precision=2, ratio_precision=3)
save_table_to_txt(df_CN, "tables/5-2 - Crank-Nicolson Table.txt", error_precision=2, ratio_precision=3)
save_table_to_txt(df_5, "tables/5-2 - Type 5 Table.txt", error_precision=2, ratio_precision=3)
save_table_to_txt(df_6, "tables/5-2 - Type 6 Table.txt", error_precision=2, ratio_precision=3)

plt.figure(figsize=(5.5, 2.5))

plt.plot(df_explicit["n"], df_explicit["Error"], color='red', marker='x', markerfacecolor='none', markeredgecolor='red', label="Explicit Euler")
plt.plot(df_implicit["n"], df_implicit["Error"], color='blue', marker='x', markerfacecolor='none', markeredgecolor='blue', label='Implicit Euler')
plt.plot(df_CN["n"], df_CN["Error"], color='purple', marker='x', markerfacecolor='none', markeredgecolor='purple', label='Crank-Nicolson')
plt.plot(df_5["n"], df_5["Error"], color='green', marker='v', markerfacecolor='none', markeredgecolor='green', label="Specular Euler (Type 5)")
plt.plot(df_6["n"], df_6["Error"], color='orange', marker='v', markerfacecolor='none', markeredgecolor='orange', label="Specular Euler (Type 6)")

plt.xscale('log', base=2)
plt.yscale('log')

plt.xlabel(r"the number of time steps $N$", fontsize=10)
plt.ylabel(r"the error ratio $\mathcal{E}(N, \infty)$", fontsize=10)
plt.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), borderaxespad=0., fontsize=10)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.savefig('figures/Numerical example 5-2.png', dpi=1000, bbox_inches='tight')

plt.show()
```

```python
t_0 = -0.2
T = 1.5
h = 0.05
max_iter = 1000

c = 0.3

def F(t, u):
    if t < 0:
        return -3*u 
    elif t > 0:
        return -3*u + 3*t + 1
    elif t == 0 and c == 1/6:
        return 1/2
    elif t == 0 and c != 1/6:
        return sd.A(1 - 3*c, -3*c) + 3*c
    
def exact_sol(t):
    t = np.asarray(t)
    return np.where(
        t < 0,
        c * np.exp(-3 * t),
        t + c * np.exp(-3 * t)
    )

def u_0(t_0):
    return exact_sol(t_0)

result_explicit = Explicit_Euler_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h) # type: ignore
result_implicit = Implicit_Euler_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h) # type: ignore
result_CN = Crank_Nicolson_Scheme(F, u_0=u_0, t_0=t_0, T=T, h=h) # type: ignore
result_5 = Specular_Euler_Scheme_Type_5(F, u_0=u_0, t_0=t_0, T=T, h=h, max_iter=max_iter) # type: ignore
result_6 = Specular_Euler_Scheme_Type_6(F, u_0=u_0, t_0=t_0, T=T, h=h, max_iter=max_iter) # type: ignore

exact_solution = np.array([exact_sol(t) for t in result_implicit[0]])

plt.figure(figsize=(5.5, 2.5))

plt.axvline(x=0.0, color='red', linestyle='--', linewidth=1)

plt.plot(result_explicit[0], exact_solution, color='black', label='exact solution') 

ymin, ymax = plt.ylim()
dy = ymax - ymin

plt.text(0.05, ymax - 0.22 * dy, "singularity", color='red', fontsize=10,
         rotation=0, ha='left', va='top',
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='red', boxstyle='round,pad=0.3'))

plt.plot(result_explicit[0], result_explicit[1],  marker='x', linestyle='None',
         markerfacecolor='none', 
         markeredgecolor='red',  
         label='Explicit Euler') 

plt.plot(result_implicit[0], result_implicit[1],  marker='x', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='blue',  
         label='Implicit Euler')  

plt.plot(result_CN[0], result_CN[1],  marker='x', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='purple',
         label='Crank-Nicolson')  

plt.plot(result_5[0], result_5[1],  marker='v', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='green',
         label='Specular Euler (Type 5)')

plt.plot(result_6[0], result_6[1],  marker='v', linestyle='None',
         markerfacecolor='none',  
         markeredgecolor='orange',  
         label='Specular Euler (Type 6)')


plt.xlabel(r"time $t$", fontsize=10)
plt.ylabel(r"solution $u(t)$", fontsize=10)
plt.grid(True)
plt.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), borderaxespad=0., fontsize=10)
plt.savefig('figures/Numerical example 7-1.png', dpi=1000, bbox_inches='tight')
plt.show()
```
